```python
#importando pacote para análise de dados
import matplotlib.pyplot as plt
import matplotlib as mat
import numpy as np
import pandas as pd
import seaborn as sns
from pandas import DataFrame
%matplotlib inline
```

```python
# carregando os dados de treino
treino = pd.read_csv(r'D:\Downloads\Titanic\train.csv')

# carregando os dados de teste
teste = pd.read_csv(r'D:\Downloads\Titanic\test.csv')
```

# Normalizando dados tabela treino

```python
treino
```

```python
#pesquisando dados faltantes na tabela treino
treino.isnull().sum()
```

```python
#preenche os dados NaN da coluna Embarcação
treino['Embarked']=treino['Embarked'].fillna('Q')
```

```python
# mudar os dados da coluna Embarcação para números
treino.loc[treino['Embarked'] == 'C', 'Embarked'] = 1
treino.loc[treino['Embarked'] == 'S', 'Embarked'] = 2
treino.loc[treino['Embarked'] == 'Q', 'Embarked'] = 3
#transforma coluna Embarcação em inteiro
treino['Embarked'] = treino['Embarked'].astype(np.int64, copy=False)
```

```python
# mudar male para 0, female para 1 tabela treino
treino.loc[treino['Sex'] == 'male', 'Sex'] = 0
treino.loc[treino['Sex'] == 'female', 'Sex'] = 1
#transforma coluna sexo em inteiro
treino['Sex'] = treino['Sex'].astype(np.int64, copy=False)
```

```python
#Excluindo colunas desnecessárias tabela treino
treino = treino.drop('Name', axis=1)
treino = treino.drop('Ticket', axis=1)
treino = treino.drop('Cabin', axis=1)
```

```python
display(treino.info())
```

```python
treino
```

# Normalizando dados tabela teste

```python
teste
```

```python
#Excluindo colunas desnecessárias tabela teste
teste = teste.drop('Name', axis=1)
teste = teste.drop('Ticket', axis=1)
teste = teste.drop('Cabin', axis=1)
```

```python
# mudar os dados da coluna Embarcação para números
teste.loc[teste['Embarked'] == 'C', 'Embarked'] = 1
teste.loc[teste['Embarked'] == 'S', 'Embarked'] = 2
teste.loc[teste['Embarked'] == 'Q', 'Embarked'] = 3
#transforma coluna Embarcação em inteiro
teste['Embarked'] = teste['Embarked'].astype(np.int64, copy=False)
```

```python
# localizando dados nulos
teste.isnull().sum()
```

```python
# localizar o dado nulo na tabela teste
teste[teste.isnull().any(axis=1)]
```

```python
# preenche os dados NaN da coluna Fare com valor médio correspondente a sua Embarcação
teste['Fare']=teste['Fare'].fillna('8.0500')
```

```python
# mudar male para 0, female para 1  tabela teste
teste.loc[teste['Sex'] == 'male', 'Sex'] = 0
teste.loc[teste['Sex'] == 'female', 'Sex'] = 1
#transforma coluna sexo e idade em inteiro
teste['Sex'] = teste['Sex'].astype(np.int64, copy=False)
```

```python
#transforma coluna Fare em float
teste['Fare'] = teste['Fare'].astype(np.float64, copy=False)
```

```python
#pesquisando dados faltantes teste
teste.isnull().sum()
```

```python
display(teste.info())
```

# Tratando idades faltantes

```python
# Aplicação de Algoritmo nas idades
```

```python
# juntando as duas tabelas, para separar as idades faltantes no geral
geral = pd.concat([treino, teste])
```

```python
display(geral.info())
```

```python
#fazendo tabela de teste para idades faltantes
#separando somente Idade com valores nulos e criando uma nova tabela
x_teste_idade_na = geral[geral['Age'].isnull()]
#excluindo a coluna Idade
x_teste_idade_na = x_teste_idade_na.drop('Age', axis = 1)

# criando uma tabela base para juntar depois
x_teste_idade_na_base = x_teste_idade_na

# excluindo survived, porque tem dados nulos
# excluindo PassengerId porque é desnecessario no treino
x_teste_idade_na = x_teste_idade_na.drop('Survived', axis = 1)
x_teste_idade_na = x_teste_idade_na.drop('PassengerId', axis = 1)
display(x_teste_idade_na.info())
```

```python
# criando x treino idade
#excluindo valores nulos e criando tabela de treino da idade
x_treino_idade_na = geral[geral['Age'].notna()]

# criar tabela treino base, para juntar depois
treino_idade_base = x_treino_idade_na

# excluindo coluna PassengerId e Survived e Idade
x_treino_idade_na = x_treino_idade_na.drop('PassengerId', axis=1)
x_treino_idade_na = x_treino_idade_na.drop('Survived', axis=1)
x_treino_idade_na = x_treino_idade_na.drop('Age', axis=1)

display(x_treino_idade_na.info())
```

```python
# criando y treino idade
y_treino_idade_na = treino_idade_base['Age']
```

```python
y_treino_idade_na
```

```python
# importando o modelo
from sklearn.linear_model import LinearRegression

# criando o modelo (objeto de regressão linear)
regr = LinearRegression()

# treinando o modelo / aplicar a regressao linear
regr.fit(x_treino_idade_na,y_treino_idade_na)
```

```python
# fazer a previsão de Idade
y_teste_idade_pred = regr.predict(x_teste_idade_na)
```

```python
#verificando valores previstos
y_teste_idade_pred
```

```python
df_idade = pd.DataFrame({'PassengerId':x_teste_idade_na_base.PassengerId,'Age':y_teste_idade_pred})
```

```python
df_idade
```

```python
# juntando as duas tabelas, predição de Idade e a tabela base
atual_idade_pred = pd.merge(x_teste_idade_na_base,df_idade, how='outer', on='PassengerId')
#transforma variavel Idade em Inteiro
atual_idade_pred['Age'] = atual_idade_pred['Age'].astype(np.int64, copy=False)
```
